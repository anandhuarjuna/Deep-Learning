```python
import tensorflow as tf
from tensorflow import keras
```

```python
import pandas as pd
df=pd.read_csv('/content/lung_cancer_examples (2).csv')


```

```python
df.isna().sum()
df.dtypes
```

```python
df.drop(['Name','Surname'],axis=1,inplace=True)
df
```

```python
X=df.iloc[:,:-1]
y=df.iloc[:,-1]
y
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
X_new=mms.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_new,y,test_size=.3,random_state=3)
```

```python
from tensorflow.keras.models import Sequential
ann=Sequential()
ann.add(keras.layers.Dense(6,activation='relu'))
ann.add(keras.layers.Dense(6,activation='relu'))
ann.add(keras.layers.Dense(1,activation='sigmoid'))

# ann.add([keras.layers.Dense(6,activation='relu'),keras.layers.Dense(6,activation='relu'),keras.layers.Dense(1,activation='sigmoid'])
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
ann.fit(X_train,y_train,epochs=100,batch_size=5)
```

```python
y_pred=ann.predict(X_test)
y_pred=(y_pred>.5)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred)*100)
```

```python
ann.summary()
```

```python
ann.evaluate(X_test,y_test)
```

```python
import warnings
warnings.filterwarnings('ignore')
y_pred1=ann.predict(mms.transform([[30,4,3,4]]))
if y_pred1>0.5:
  print('chances of lungh cancer is high')
else:
  print('chances of lungh cancer is low')
```
